### Add two columns High_Low_Speed and High_Low_HP in the dataset of Pokemon

```python
import pandas as pd
```

```python
data=pd.read_csv("Pokemon.csv")
```

```python
data.head()
```

### Adding Column High_Low_Speed


```python
Speed_mean=data["Speed"].mean() 
Speed_mean
```

```python
Speed_mean=data["Speed"].mean()  

def set_speed(val):
    if val < Speed_mean:
        return "Low Speed"
    elif val == Speed_mean:
        return "Neutral Speed"
    else:
        return "High Speed"
    
```

```python
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data
```

### Adding Column High_Low_HP


```python
HP_mean=data["HP"].mean()  
HP_mean
```

```python
HP_mean=data["HP"].mean()  

def set_HP(val):
    if val < HP_mean:
        return "Low HP"
    elif val == HP_mean:
        return "Neutral HP"
    else:
        return "High HP"
    
```

```python
data["High_Low_HP"] = data["HP"].apply(set_HP)
data
```
